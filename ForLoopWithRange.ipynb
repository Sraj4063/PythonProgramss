```python
''' Given two integers A and B (A ≤ B). Print all numbers from A to B inclusively. '''
a = int(input())
b = int(input())
if (a == b):
    print(a)
else:
    for i in range(a,b+1):
        print(i)
        
# or

a = int(input())
b = int(input())
for i in range(a, b + 1):
    print(i)
```

```python
''' Given two integers A and B. Print all numbers from A to B inclusively, in ascending order, if A < B, or in 
descending order, if A ≥ B.  '''
a = int(input())
b = int(input())
if a < b:
    for i in range(a, b + 1):
        print(i)
else:
    for i in range(a, b - 1, -1):
        print(i)

# or

a = int(input())
b = int(input())
if a < b:
    for i in range(a, b + 1):
        print(i)
else:
    for i in range(a, b - 1, -1):
        print(i)
```

```python
'''10 numbers are given in the input. Read them and print their sum. Use as few variables as you can.'''
sum = 0
i = 1
while (i <= 10):
    sum += int(input())
    i = i+1
print(sum)  

# or

res = 0
for i in range(10):
    res += int(input())
print(res)
```

```python
''' N numbers are given in the input. Read them and print their sum.
The first line of input contains the integer N, which is the number of integers to follow. Each of the next N lines
contains one integer. Print the sum of these N integers. '''

n = int(input())
sum = 0
i = 0
while (i < n):
    sum += int(input())
    i += 1
print(sum)    

# or

n = int(input())
res = 0
for i in range(n):
    res += int(input())
print(res)
```

```python
''' For the given integer N calculate the following sum:
1^3+2^3+…+N^3
'''
n = int(input())
sum = 0
for i in range(0, n+1):
    sum += i**3
print(sum)    

# or

res = 0
for i in range(1, int(input()) + 1):
    res += i ** 3
print(res)
```

```python
''' In mathematics, the factorial of an integer n, denoted by n! is the following product:
n!=1×2×…×n

For the given integer n
calculate the value n!. Don't use math module in this exercise. '''
a = int(input())
sum = 1
if (a== 1):
    print(a)
else:
    for i in range(2,a+1):
        sum *= i
    print(sum)    
    
# or

res = 1
n = int(input())
for i in range(1, n + 1):
    res *= i
print(res)
```

```python
''' Given N numbers: the first number in the input is N, after that N integers are given. Count the number of zeros 
among the given integers and print it.
You need to count the number of numbers that are equal to zero, not the number of zero digits.
'''
n = int(input())
count = 0;
for i in range(1, n+1):
    a = int(input())
    if (a == 0):
        count += 1
print(count)        

# or

num_zeroes = 0
for i in range(int(input())):
    if int(input()) == 0:
        num_zeroes += 1
print(num_zeroes)

```

```python
''' Given an integer n, print the sum 1!+2!+3!+...+n!.
This problem has a solution with only one loop, so try to discover it. And don't use the math library :) '''
n = int(input())
sum = 1
temp = 1

for i in range(n,1, -1):
    for j in range(i,1,-1):
        temp *= j;
    sum += temp
    temp = 1
print(sum)    

# or

n = int(input())
partial_factorial = 1
partial_sum = 0
for i in range(1, n + 1):
    partial_factorial *= i
    partial_sum += partial_factorial
print(partial_sum)
```

```python
'''For given integer n ≤ 9 print a ladder of n steps. The k-th step consists of the integers from 1 to k without 
spaces between them.
To do that, you can use the sep and end arguments for the function print(). '''

n = int(input())
for i in range(1, n+1):
    for j in range(1,i+1):
        print(j,end = "")
    print()    
    
# or

n = int(input())
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, sep='', end='')
    print()
```

```python
'''There was a set of cards with numbers from 1 to N. One of the card is now lost. Determine the number on that lost card given the numbers for the remaining cards.

Given a number N, followed by N − 1 integers - representing the numbers on the remaining cards (distinct integers in the range from 1 to N). Find and print the number on the lost card. '''

lst = []
n = int(input())
for i in range(1,n):
    a = int(input())
    lst.append(a)
for i in range(1,n+1):
    if i not in lst:
        print(i)
        
# or

n = int(input())
sum_cards = 0
for i in range(1, n + 1):
    sum_cards += i
# One can prove the following:
# sum_cards == n * (n + 1) // 2
# However, we'll calculate that using the loop.
for i in range(n - 1):
    sum_cards -= int(input())
print(sum_cards)

```
