```python
''' For a given integer N, print all the squares of integer numbers where the square is less than or equal to N, 
in ascending order. '''
n = int(input())
i = 1
while i ** 2 <= n:
    print(i ** 2)
    i += 1
```

```python
'''For a given integer N, find the greatest integer x where 2x is less than or equal to N. Print the exponent value 
and the result of the expression 2x.
Don't use the operation **.'''

n = int(input())
i =0
for i in range(0,n+1):
    if 2**i <= n:
        i += 1
        if 2**i > n:
            i -= 1
            break
print(i,2**i)            

# or

n = int(input())
two_in_power = 2
power = 1
while two_in_power <= n:
    two_in_power *= 2
    power += 1
print(power - 1, two_in_power // 2)
```

```python
''' As a future athlete you just started your practice for an upcoming event. Given that on the first day you run x
miles, and by the event you must be able to run y miles.
Calculate the number of days required for you to finally reach the required distance for the event, if you increases 
your distance each day by 10% from the previous day.
Print one integer representing the number of days to reach the required distance. '''

x = int(input("How many miles you run now? "))
y = int(input("How many miles you must be able to run? "))
count = 1

while x < y:
  x += x * 0.1
  count += 1
    
print("The number of days to reach the required distance: " + str(count))

# or
x = int(input())
y = int(input())
i = 1
while x < y:
    x *= 1.1
    i += 1
print(i)
```
